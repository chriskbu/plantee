```python
import json
import requests
import xmltodict
import sqlite3
```

```python
# Database Stuff
db_conn = sqlite3.connect("weather.db")  #Create database
cursor = db_conn.cursor() # Create db cursor
```

```python
# Create table
cursor.execute('''CREATE TABLE weather_data (area_name text, date text, temperature real, percipitation real)''') # Create table in db
```

```python
cursor.execute('''INSERT INTO weather_data VALUES('vestgrensa', '08-11-93', -10.0, 0.0)''') # Insert random data
```

```python
db_conn.commit() # Commit changes to db
```

```python
db_reading = cursor.execute('''SELECT * FROM weather_data''') # Query the db

# Print results from query
for row in db_reading:
    print(row)
```

```python
r = requests.get("https://www.yr.no/place/Norway/Oslo/Oslo/Vestgrensa~2254728/varsel.xml")
r.status_code
```

```python
a = xmltodict.parse(r.content)
```

```python
print(json.dumps(a, sort_keys=True, indent=4))
```

```python
sun_rise = a["weatherdata"]["sun"]["@rise"].split("T")
sun_set = a["weatherdata"]["sun"]["@set"].split("T")
```
